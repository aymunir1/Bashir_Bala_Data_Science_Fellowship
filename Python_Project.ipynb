Feature Ene

```python
#Conversion of the catagorical columns to numerical columns

brcad["Gender"] = brcad["Gender"].map({"FEMALE": 1, "MALE":0})
brcad["Tumour_Stage"] = brcad["Tumour_Stage"].map({"I": 1, "II":2, "III":3, "IV":4})
brcad["Histology"] = brcad["Histology"].map({"Infiltrating Ductal Carcinoma": 1, "Infiltrating Labular Carcinoma":2,  "Mucinous Carcinoma":3})
brcad["ER status"] = brcad["ER status"].map({"Positive": 1, "Negative":0})
brcad["PR status"] = brcad["PR status"].map({"Positive": 1, "Negative":0})
brcad["HER2 status"] = brcad["HER2 status"].map({"Positive": 1, "Negative":0})
brcad["Surgery_type"] = brcad["Surgery_type"].map({"Lumpectomy": 1, "Modified Radical Mastectomy":2, "Simple Masterctomy":3, "Other":4})
brcad["Patient_Status"] = brcad["Patient_Status"].map({"Alive": "Alive", "Dead":"Dead"})
brcad["Age_Group"] = brcad["Age_Group"].map({"0-29":1, "30-39":2, "40-49":3, "50-59":4, "60-69":5, "70-79":6, "80-89":7, "90-99":8})

brcad.head()
```

```python
brcad= brcad.dropna()
```

```python
  #splitling the data to train and test
  x= brcad.drop(['Patient_ID','Patient_Status','Date_of_Surgery',  'Date_of_Last_Visit'], axis=1)
y= brcad['Patient_Status']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Training the model
model = SVC()


# Fitting data into the model
model.fit(x_train, y_train)
```

```python
#Making Prediction
Prediction =np.array(model.predict([[35,2,0.213900,0.42638, 0.124270, 0.52045,2,1,2,1,2,1,3]]))

print(Prediction)
```

```python
# Evaluating the Model
evaluation = model.score(x_train, y_train)
print(evaluation)
```
